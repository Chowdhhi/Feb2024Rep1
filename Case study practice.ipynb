```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
kd=pd.read_csv(r'C:\Users\hindo\Downloads\kidney_disease.csv')
kd
```

```python
dc=kd.isna().sum()/len(kd)*100
dci=dc[dc>25].index
```

```python
kd.drop(dci,axis=1,inplace=True)
```

```python
kd.head()
```

```python
kd.drop('id',axis=1,inplace=True)
```

```python
kd.head()
```

```python
num_lst=[]
for i in kd:
    if kd[i].dtype!=object:
        num_lst.append(i)
```

```python
def null(col):
    return kd[col].fillna(kd[col].mean(),inplace=True)
```

```python
for i in num_lst:
    null(i)
```

```python
obj_lst=[]
for i in kd:
    if kd[i].dtype==object:
        obj_lst.append(i)
```

```python
def cat_null(col):
    return kd[col].fillna(kd[col].mode()[0],inplace=True)
```

```python
for i in obj_lst:
    cat_null(i)
```
