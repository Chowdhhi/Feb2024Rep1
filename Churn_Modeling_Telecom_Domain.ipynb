```python
def churn_details(dataset, churn_column='Churn'):
    # Filter the dataset for customers who are churning
    churned_customers = dataset[dataset[churn_column] == 'Yes']

    # Create a DataFrame to store details for each column
    details_df = pd.DataFrame(columns=['Column', 'Unique_Values', 'Missing_Values'])

    # Iterate through each column
    for column in dataset.columns:
        unique_values = dataset[column].nunique()
        missing_values = dataset[column].isnull().sum()
        
        # Store details in the DataFrame
        details_df = details_df.append({
            'Column': column,
            'Unique_Values': unique_values,
            'Missing_Values': missing_values
        }, ignore_index=True)

    return details_df
```

```python
details_for_churning_customers = churn_details(dataset)

# Displaying the details
print(details_for_churning_customers)
```

```python
# split the data into training and test

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size=0.25, random_state=44)
```

# RandomForest Classifier

```python
#pip install scikit-learn
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(oob_score=True)
rf.fit(x_train, y_train)
```

```python
y_pred_rf_train = rf.predict(x_train)
y_pred_rf_test = rf.predict(x_test)
```

```python
# Evaluation matrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_train, y_pred_rf_train))
print()
print(confusion_matrix(y_test, y_pred_rf_test))
```

```python
print(classification_report(y_train, y_pred_rf_train))
print()
print(classification_report(y_test, y_pred_rf_test))
```

```python
print("Training Accuracy :",accuracy_score(y_train, y_pred_rf_train))
print()
print("Test Accuracy :", accuracy_score(y_test, y_pred_rf_test))
```

```python
# Cross Validation - K_Fold method
from sklearn.model_selection import cross_val_score
training_accuracy = cross_val_score(rf, x_train, y_train, cv=10)
print(training_accuracy)
print()
print(training_accuracy.mean())
print()
print(training_accuracy.max())
```
