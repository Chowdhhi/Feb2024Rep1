```python
def get_missing_data_details1(df):
    # Create a heatmap to visualize missing values
    plt.figure(figsize=(10, 5))
    sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
    plt.title('Missing Data: Heatmap', fontsize=15)
    plt.show()

    # Calculate and plot missing percentage for each feature
    total = df.isnull().sum().sort_values(ascending=False)
    percent = ((df.isnull().sum() / df.isnull().count()) * 100).sort_values(ascending=False)
    missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    missing = missing[missing['Percent'] > 0]

    plt.figure(figsize=(10, 5))
    sns.barplot(x=missing.index, y=missing['Percent'], palette='viridis')  # Switch x and y
    plt.xlabel('Features', fontsize=12)
    plt.ylabel('Percentage of Missing Data', fontsize=12)
    plt.title('Percentage of Missing Data by Feature', fontsize=15)
    plt.show()
```

```python
def analyse_columnwise_data(data):
    ana = data.str.findall('[a-zA-Z]+',)
    ana = [x[0] for x in ana if x is not np.nan]
    return set(ana)

def get_brand(s):
    return s.split()[0]
def get_model(s):
    return s.split()[1]
```

```python
def convert_new_price(s):
    if type(s)==str:
        s = s.lower()
        if 'lakh' in s:
            s = s.strip("lakh")
            s = float(s)
        elif "cr" in s:
            s = s.strip("cr")
            s = float(s)*100
        return s
```

```python
train_data['New_Price_Converted'] = train_data['New_Price'].apply(convert_new_price)
```

```python
train_data.to_csv('new_car.csv')
```

```python
def adjust_mileage(s):
    if type(s) == str:
        s = s.lower()
        if 'kmpl' in s:
            s = s.strip('kmpl')
        if 'km/kg' in s:
            s = s.strip('km/kg')
            s = float(s)*1.8
    return s
```
